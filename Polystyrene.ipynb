```python
def get_T_unc(all_T, all_dT):

    all_T = np.array(all_T)
    all_dT = np.array(all_dT)
    unc = np.sqrt(np.std(all_T)**2 + np.mean(all_dT**2))
    return unc
```

```python
def get_k_unc(all_k, all_dk):

    all_k = np.array(all_k)
    all_dk = np.array(all_dk)
    unc = np.sqrt(np.std(all_k)**2 + np.mean(all_dk**2))
    return unc
```

```python
data = np.load('PS_deform_40_monomers.npz')
ep=data['ep']
strains = data['strains']
k= data['k']
dk= data['dk']
T= data['T']
tacticity= data['tacticity']
```

```python
cmap = mpl.colormaps.get_cmap('Spectral_r')
norm = mpl.colors.Normalize(np.min(T), np.max(T))
sm = mpl.cm.ScalarMappable(cmap=cmap, norm=norm)

all_markers = ['^', 'd', 's', 'o', 'h', 'v']
size = 10

for i, t in enumerate(tacticity):

    if i == 0:
        ep_0 = ep[i, ~np.isnan(k[i])]
        k_0 = k[i, ~np.isnan(k[i])]
        T_0 = k[i, ~np.isnan(k[i])]
        plt.plot(ep_0, k_0, f"{all_markers[i]}--", markerfacecolor='white', color='black', markersize=size, label=f'{t}%')
        for j in range(len(ep_0)):
                    plt.plot(ep_0[j], k_0[j], f"{all_markers[i]}", markerfacecolor=cmap(norm(T_0[j])), markeredgecolor='black', markersize=size)
        continue
    else:
        plt.plot(ep[i], k[i], f"{all_markers[i]}--", markerfacecolor='white', color='black', markersize=size, label=f'{t}%')
        for j in range(len(ep[i])):
                    plt.plot(ep[i,j], k[i,j], f"{all_markers[i]}", markerfacecolor=cmap(norm(T[i,j])), markeredgecolor='black', markersize=size)

plt.colorbar(sm, label='Chain Tension (pN)', ax=plt.gca())
plt.legend()
plt.ylabel('Thermal Conductivity (W/m-K)')
plt.xlabel('Strain (%)')
plt.yscale('log')
plt.xlim(-12, 12)
plt.ylim(1,30)
# plt.title('Polystyrene')
plt.tight_layout()
# plt.savefig('PS_strain.svg', format='svg')
plt.show()
```
