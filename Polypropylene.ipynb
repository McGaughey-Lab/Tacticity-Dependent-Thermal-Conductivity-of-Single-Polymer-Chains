```python
def get_T_unc(all_T, all_dT):

    all_T = np.array(all_T)
    all_dT = np.array(all_dT)
    unc = np.sqrt(np.std(all_T)**2 + np.mean(all_dT**2))
    return unc
```

```python
def get_k_unc(all_k, all_dk):

    all_k = np.array(all_k)
    all_dk = np.array(all_dk)
    unc = np.sqrt(np.std(all_k)**2 + np.mean(all_dk**2))
    return unc
```

```python
with open(f'PP_dihedral_40.pickle', 'rb') as f:
    data = pickle.load(f)

d_all_0 = data["0"]["Dihedrals"]
d_all_10 = data["10"]["Dihedrals"]
d_all_20 = data["20"]["Dihedrals"]
d_all_30 = data["30"]["Dihedrals"]
d_all_40 = data["40"]["Dihedrals"]
d_all_50 = data["50"]["Dihedrals"]
d_all_60 = data["60"]["Dihedrals"]
d_all_70 = data["70"]["Dihedrals"]
d_all_80 = data["80"]["Dihedrals"]
d_all_90 = data["90"]["Dihedrals"]
d_all_100 = data["100"]["Dihedrals"]

b = 0.25
kde_0 = gaussian_kde(d_all_0, bw_method=b)
kde_10 = gaussian_kde(d_all_10, bw_method=b)
kde_20 = gaussian_kde(d_all_20, bw_method=b)
kde_30 = gaussian_kde(d_all_30, bw_method=b)
kde_40 = gaussian_kde(d_all_40, bw_method=b)
kde_50 = gaussian_kde(d_all_50, bw_method=b)
kde_60 = gaussian_kde(d_all_60, bw_method=b)
kde_70 = gaussian_kde(d_all_70, bw_method=b)
kde_80 = gaussian_kde(d_all_80, bw_method=b)
kde_90 = gaussian_kde(d_all_90, bw_method=b)
kde_100 = gaussian_kde(d_all_100, bw_method=b)

angle = np.linspace(0, 360, 361, endpoint=True)
d0 = kde_0(angle)
d10 = kde_10(angle)
d20 = kde_20(angle)
d30 = kde_30(angle)
d40 = kde_40(angle)
d50 = kde_50(angle)
d60 = kde_60(angle)
d70 = kde_70(angle)
d80 = kde_80(angle)
d90 = kde_90(angle)
d100 = kde_100(angle)

cmap = mpl.colormaps.get_cmap('autumn')
norm = mpl.colors.Normalize(0, 100)
sm = mpl.cm.ScalarMappable(cmap=cmap, norm=norm)

# Plot each graph, and manually set the y tick values
plt.figure(figsize=(5,5))

plt.plot(angle, d20, label='20%', color=cmap(norm(20)))
plt.plot(angle, d40, label='40%', color=cmap(norm(40)))
plt.plot(angle, d60, label='60%', color=cmap(norm(60)))
plt.plot(angle, d80, label='80%', color=cmap(norm(80)))

plt.plot(angle, d0, label='0%', color='black')
plt.plot(angle, d100, label='100%', color='purple')


plt.xlabel('Dihedral Angle (degrees)')
plt.ylabel('Probability')
plt.title('Polypropylene (40 Monomers)')
plt.legend()
plt.tight_layout()

plt.xlim(0,360)
plt.ylim(0, 0.025)

# plt.savefig('PP_dihedral_40.svg', format='svg')

plt.show()
```
